```python
from astropy.cosmology import FlatLambdaCDM
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
distance_cluster = cosmo.angular_diameter_distance(z=0.228)
print(distance_cluster)
len_1arc = distance_cluster*u.arcsec.to(u.rad)
len_pix = len_1arc * 0.322

velocity_dispersion = 1100 * u.km/u.s
v_vir = velocity_dispersion.value
# v_vir_los = v_vir/np.sqrt(3) # los vel dispersion
R_vir = 1.51/0.7  # Mpc
l = len_pix.value/R_vir # 1 pix = l R_vir

len_pix
```

```python
def draw_phase_diagram(Table, r1=1., v1=1., r2=2., v2 = 3., z0=0.228, ax=None):
    Tab = Table.groupby("ID", sort=False)   
    d_centroid_v = Tab.diff_centroid.median()
    d_angle_v = Tab.diff_angle.median()
    
    cond1 =  (d_centroid_v>1)&(d_angle_v<90)
    cond2 = (d_centroid_v>1)&(d_angle_v>90)
    cond1b =  (d_centroid_v>.5)&(d_angle_v<90)
    cond2b = (d_centroid_v>.5)&(d_angle_v>90)

    R_norm = Tab.dist_cen.median()*l
    v_norm = abs(Tab.redshift.median()-z0)*3e5/v_vir
    mag = Tab.mag_auto.median()

    from matplotlib.patches import Ellipse
    if ax is None:
        plt.figure(figsize=(11,9))
        ax = plt.subplot(111)
    # (25.2-mag[cond1])**2+20
    plt.scatter(R_norm[cond2], v_norm[cond2], s=150, color="firebrick",
                edgecolor="none",alpha=0.7,zorder=3,label="$\Delta\,d>1$ pix , $\\theta>90^\circ$")
    plt.scatter(R_norm[cond1], v_norm[cond1], s=150, color="steelblue",
                edgecolor="none",alpha=0.7,zorder=3,label="$\Delta\,d>1$ pix , $\\theta<90^\circ$")
#     plt.scatter(R_norm[cond1b], v_norm[cond1b], s=50, 
#                 facecolor="None",color="steelblue",lw=1.2,
#                 alpha=0.7,zorder=4,label="$\Delta\,d > 0.5$ pix , $\\theta<90^\circ$")
#     plt.scatter(R_norm[cond2b], v_norm[cond2b], s=50, 
#                 facecolor="None",color="orange",lw=1.2,
#                 alpha=0.7,zorder=4,label="$\Delta\,d > 0.5$ pix , $\\theta>90^\circ$")

    non_virial = (((R_norm/r1)**2 + (v_norm/v1)**2) > 1.) & (((R_norm/r2)**2 + (v_norm/v2)**2) < 1.)
    virial = (((R_norm/r1)**2 + (v_norm/v1)**2) < 1.)
    inner = (R_norm/r1)**2 < 1.
    outer = (R_norm/r1)**2 > 1.
    
    condA = non_virial & (R_norm<0.5*v_norm)
    condB = virial
    condC = non_virial & (R_norm>0.5*v_norm)

    condA1, condA1b = condA&cond1, condA&cond1b 
    condB1, condB1b = condB&cond1, condB&cond1b 
    condC1, condC1b = condC&cond1, condC&cond1b 
    condA2, condA2b = condA&cond2, condA&cond2b 
    condB2, condB2b = condB&cond2, condB&cond2b 
    condC2, condC2b = condC&cond2, condC&cond2b 
    plt.text(0.05, 0.84,r"$\bf A$",color="k",va="center",fontsize=25,transform=ax.transAxes)
    plt.text(0.05, 0.22,r"$\bf B$",color="k",va="center",fontsize=25,transform=ax.transAxes)
    plt.text(0.85, 0.1,r"$\bf C$",color="k",va="center",fontsize=25,transform=ax.transAxes)
    
    ft = 22
    plt.text(0.62, 0.94,"A :",color="k",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.62, 0.85,"B :",color="k",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.62, 0.76,"C :",color="k",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.7, 0.94,"%d"%(np.sum(condA1)),color="steelblue",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.7, 0.85,"%d"%(np.sum(condB1)),color="steelblue",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.7, 0.76,"%d"%(np.sum(condC1)),color="steelblue",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.86, 0.94,"%d"%(np.sum(condA2)),color="firebrick",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.86, 0.85,"%d"%(np.sum(condB2)),color="firebrick",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.86, 0.76,"%d"%(np.sum(condC2)),color="firebrick",va="center",fontsize=ft,transform=ax.transAxes)
    plt.text(0.74, 0.94,"(%d)"%(np.sum(condA1b)),color="steelblue",va="center",fontsize=ft-2,transform=ax.transAxes)
    plt.text(0.74, 0.85,"(%d)"%(np.sum(condB1b)),color="steelblue",va="center",fontsize=ft-2,transform=ax.transAxes)
    plt.text(0.74, 0.76,"(%d)"%(np.sum(condC1b)),color="steelblue",va="center",fontsize=ft-2,transform=ax.transAxes)
    plt.text(0.9, 0.94,"(%d)"%(np.sum(condA2b)),color="firebrick",va="center",fontsize=ft-2,transform=ax.transAxes)
    plt.text(0.9, 0.85,"(%d)"%(np.sum(condB2b)),color="firebrick",va="center",fontsize=ft-2,transform=ax.transAxes)
    plt.text(0.9, 0.76,"(%d)"%(np.sum(condC2b)),color="firebrick",va="center",fontsize=ft-2,transform=ax.transAxes)

    ellipse1 = Ellipse(xy=(0,0), width=2*r1, height=2*v1, 
                      edgecolor='gray', fc='gray', lw=3, alpha=0.5,zorder=2)
    ellipse2 = Ellipse(xy=(0,0), width=2*r2, height=2*v2, 
                      edgecolor='gray', fc='gray', lw=3, alpha=0.25,zorder=1)
    ax.add_patch(ellipse1)
    ax.add_patch(ellipse2)
    xx = np.linspace(0.45*r1,0.6*r2,20)
    plt.plot(xx,2*xx,c="gray",ls="--",lw=3, alpha=0.5,zorder=2)

    plt.xlim(-0.005, r2+0.05)
    plt.ylim(-0.005, v2+0.1)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.xlabel("r / R$_{200}$",fontsize=15)
    plt.ylabel("|v$_{los}$ - v$_0$| / $\sigma_{vir}$",fontsize=15)
#     plt.legend(loc=9,fontsize=14)
    plt.tight_layout()
    return (virial, non_virial, inner, outer)
    # plt.savefig("/home/qliu/Pictures/Phase_A2390.pdf",dpi=150)    
```
